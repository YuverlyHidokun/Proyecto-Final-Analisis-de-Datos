```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar datos desde un archivo CSV utilizando pandas
data = pd.read_csv('Concerts_Taylor_Clean.csv')
```

```python
# Seleccionar las primeras 30 filas para el gráfico
data_x = data
data_y = data

# Extraer datos para los ejes X e Y
x = data_x['Tour']
y = data_y['Attendance (tickets sold)']

# Crear un gráfico de barras
plt.bar(x, y)
plt.title('Tickets vendidos por tour')
plt.xlabel('Tour')
plt.ylabel('Attendance (tickets sold)')
plt.xticks(rotation=90, ha="right")  # Rotar las etiquetas del eje X para mayor legibilidad
plt.show()
```

```python
# Seleccionar las primeras 5 filas para el gráfico de pastel (por ejemplo)
data_subset = data.head(10)

# Extraer datos para los ejes X e Y
labels = data_subset['City']
sizes = data_subset['Revenue']

# Crear un gráfico de pastel
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Distribución de Tickets vendidos por tour')
plt.axis('equal')  # Hace que el gráfico de pastel sea circular

plt.show()
```

```python
data_subset = data.head(10)

# Extraer datos para los ejes X e Y
x = data_subset['City']
y = data_subset['Attendance (available)']

# Crear un gráfico de líneas
plt.plot(x, y, marker='o', linestyle='-', color='b', label='Tickets disponibles por ciudad')

# Personalizar el gráfico
plt.title('Tickets disponibles por ciudad')
plt.xlabel('City')
plt.ylabel('Attendance (available)')
plt.xticks(rotation=90, ha="right")
plt.legend()

plt.show()
```
