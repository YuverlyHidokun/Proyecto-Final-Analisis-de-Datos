```python
import pandas as pd
import mysql.connector

# Configuración de la conexión a la base de datos
host = 'localhost'
user = 'root'
password = '123456'
database = 'conciertos'

# Crear la conexión a la base de datos
conn = mysql.connector.connect(host=host, user=user, password=password, database=database)

# Consulta SQL para obtener datos de la tabla
query = 'SELECT * FROM conciertos_taylor'

# cargar los resultados en un df
df = pd.read_sql(query, conn)


conn.close()

print(df)
```

```python
# Mostrar las primeras filas del dataset
print(df.head())

# Obtener información sobre las columnas
print(df.info())

# Estadísticas descriptivas del dataset
print(df.describe())
```

```python
# Eliminar filas con valores nulos
df = df.dropna()
```

```python
# Reemplazar los valores nulos con el valor más repetido en esa columna
valor_mas_repetido = df['City'].mode()[0]
df['City'].fillna(valor_mas_repetido, inplace=True)
```

```python
valor_mas_repetido = df['Country'].mode()[0]
df['Country'].fillna(valor_mas_repetido, inplace=True)
```

```python
valor_mas_repetido = df['Venue'].mode()[0]
df['Venue'].fillna(valor_mas_repetido, inplace=True)
```

```python
valor_mas_repetido = df['Opening act(s)'].mode()[0]
df['Opening act(s)'].fillna(valor_mas_repetido, inplace=True)
```

```python
valor_mas_repetido = df['Attendance (tickets sold / available)'].mode()[0]
df['Attendance (tickets sold / available)'].fillna(valor_mas_repetido, inplace=True)
```

```python
valor_mas_repetido = df['Revenue'].mode()[0]
df['Revenue'].fillna(valor_mas_repetido, inplace=True)
```

```python
valor_mas_repetido = df['Tour'].mode()[0]
df['Tour'].fillna(valor_mas_repetido, inplace=True)
```

```python
# Eliminar duplicados
df = df.drop_duplicates()
```

```python
# columna que contiene valores con el símbolo de dólar
df['Revenue'] = df['Revenue'].str.replace('$', '')
```

```python
# columna que contiene valores con comas
df['Revenue'] = df['Revenue'].str.replace(',', '')
```

```python
# columna que contiene valores con caracteres no deseados
df['Revenue'] = df['Revenue'].replace('[^\d.]', '', regex=True)
```

```python
df['Revenue'] = df['Revenue'].astype(int)
```

```python
# columna que contiene valores en formato string
valor_por_defecto = 0

# Reemplazar cadenas vacías con el valor por defecto 
#antes de la conversión a enteros
df['Revenue'] = df['Revenue'].replace('', valor_por_defecto)

# Convertir la columna a enteros
df['Revenue'] = df['Revenue'].astype(int)

print(df)
```

```python
df['Revenue'].replace(0, pd.NA, inplace=True)

# Encontrar el valor más repetido
valor_mas_repetido = df['Revenue'].mode()[0]

# Reemplazar ceros con el valor más repetido y NaN con 0
df['Revenue'] = df['Revenue'].replace({0: valor_mas_repetido, pd.NA: 0})

print(df)
```

```python
df['Revenue'].replace(0, pd.NA, inplace=True)

# Encontrar el valor más repetido
valor_mas_repetido = df['Revenue'].mode()[0]

# Reemplazar ceros con el valor más repetido
df['Revenue'].replace(pd.NA, valor_mas_repetido, inplace=True)

print(df)
```

```python
# Crear una copia del DataFrame original
df_copy = df.copy()
```

```python
# Creamos nuevas columnas 'Attendance (tickets sold)' y 'Attendance (available)'
df_copy[['Attendance (tickets sold)', 'Attendance (available)']] = df_copy['Attendance (tickets sold / available)'].str.split('/', expand=True)

# Eliminamos la columna original 'Attendance (tickets sold / available)'
df_copy.drop(columns=['Attendance (tickets sold / available)'], inplace=True)

# Eliminamos los caracteres "," de las nuevas columnas
df_copy['Attendance (tickets sold)'] = df_copy['Attendance (tickets sold)'].str.replace(',', '')
df_copy['Attendance (available)'] = df_copy['Attendance (available)'].str.replace(',', '')

# Reemplazamos valores nulos, vacíos, con cero en las nuevas columnas
df_copy['Attendance (tickets sold)'] = df_copy['Attendance (tickets sold)'].fillna(0)
df_copy['Attendance (available)'] = df_copy['Attendance (available)'].fillna(0)

print(df_copy)
```

```python
# Eliminar los caracteres "," de las nuevas columnas
df_copy['Attendance (tickets sold)'] = df_copy['Attendance (tickets sold)'].str.replace(',', '')

# Reemplazar el carácter "�" con cero en 'Attendance (tickets sold)'
df_copy['Attendance (tickets sold)'] = df_copy['Attendance (tickets sold)'].replace('�', '0')

# Reemplazar valores nulos, vacíos y None con cero en 'Attendance (tickets sold)'
df_copy['Attendance (tickets sold)'] = df_copy['Attendance (tickets sold)'].fillna(0)
```

```python
# Convertir las columnas a tipo de datos entero
df_copy['Attendance (tickets sold)'] = df_copy['Attendance (tickets sold)'].astype(int)
df_copy['Attendance (available)'] = df_copy['Attendance (available)'].astype(int)
```

```python
# Encontrar el valor más repetido excluyendo ceros en 'Attendance (tickets sold)' y 'Attendance (available)'
valor_tickets_repetido = df_copy['Attendance (tickets sold)'][df_copy['Attendance (tickets sold)'] != 0].mode().iloc[0]
valor_available_repetido = df_copy['Attendance (available)'][df_copy['Attendance (available)'] != 0].mode().iloc[0]

# Reemplazar ceros con el valor más repetido excluyendo ceros
df_copy['Attendance (tickets sold)'] = df_copy['Attendance (tickets sold)'].replace(0, valor_tickets_repetido)
df_copy['Attendance (available)'] = df_copy['Attendance (available)'].replace(0, valor_available_repetido)
```

```python
df_copy['Opening act(s)'] = df_copy['Opening act(s)'].fillna('Gloriana Kellie Pickler')
```
