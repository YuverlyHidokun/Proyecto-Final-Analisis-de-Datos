```python
import matplotlib.pyplot as plt
import pandas as pd
from pymongo import MongoClient
mongo_url = "mongodb://localhost:27017/MenuMc"
client = MongoClient(mongo_url)
db = client.MenuMc  # Base de datos
collection = db.MenuMc  # Colección
documentos = list(collection.find())
df = pd.DataFrame(documentos)
# Gráfico de barras de calorías promedio por categoría
promedio_calorias = df.groupby('Category')['Calories'].mean()
promedio_calorias.plot(kind='bar')
plt.ylabel('Calorías promedio')
plt.show()
```

```python
# Gráfico de correlación de nutrientes
nutrientes = df[['Total Fat', 'Cholesterol', 'Sodium', 'Carbohydrates', 'Dietary Fiber', 'Sugars', 'Protein']]
correlacion = nutrientes.corr()
plt.imshow(correlacion, cmap='hot', interpolation='none')
plt.colorbar()
plt.xticks(range(len(correlacion)), correlacion.columns, rotation=90)
plt.yticks(range(len(correlacion)), correlacion.columns)
plt.gcf().set_size_inches(8, 8)
plt.show()
```
