```python
#Limpieza del csv
import pandas as pd

nombre_archivo = 'cause_of_deaths.csv'
datos = pd.read_csv(nombre_archivo)

datos_limpios = datos.fillna(0)


nombre_archivo_limpio = 'cause_of_deaths Limpio.csv'
datos_limpios.to_csv(nombre_archivo_limpio, index=False)

print("Limpieza de datos completa. El archivo limpio se ha guardado como:", nombre_archivo_limpio)
```

```python
#Analisis de los Datos
import pandas as pd
import matplotlib.pyplot as plt

def analizar_datos_csv(ruta_archivo):
    datos = pd.read_csv(ruta_archivo)

    datos['Malaria'].plot(kind='hist', bins=20)  
    plt.title('Histograma de muertes por malaria en el planeta')
    plt.xlabel('Malaria')
    plt.ylabel('Year')
    plt.show()


ruta_archivo = 'cause_of_deaths.csv'
analizar_datos_csv(ruta_archivo)
```

```python
import matplotlib.pyplot as plt

def grafico_pastel(etiquetas, sizes):
    colores = ['gold', 'yellowgreen']
    
    plt.figure(figsize=(6,6))
    plt.pie(sizes, labels=etiquetas, colors=colores, autopct='%1.1f%%', startangle=140)
    
    plt.title('las muertes mas comunes del Ecuador')

    plt.axis('equal')  
    plt.show()
etiquetas = ['Meningitis','Malaria']
sizes = [25, 35]
grafico_pastel(etiquetas, sizes)
```
